Based on the classification(light rain, heavy rain and violent rain), using the rain column, I made another column indicating the type of rainfall

```python
Rain_type = []
for row in df1['rain']:
  if row < 1 : Rain_type.append('no_rain')
  elif row < 2.6 : Rain_type.append('light_rain')
  elif row < 7.6 : Rain_type.append('moderate_rain')
  elif row < 51 : Rain_type.append('heavy_rain')
  elif row > 50 : Rain_type.append('Violent_rain') 
```

```python
df1['Rain_type'] = Rain_type
```
